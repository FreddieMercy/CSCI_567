```python
def phi(x):
    return (x, x**2)
```

```python
data = {"x1":[], "x2":[], "label":[]}
features = [-1, 0, 1]
labels = [-1,1,-1]

for i in range(len(features)):
    points = phi(features[i])
    data["x1"].append(points[0])
    data["x2"].append(points[1])
    data["label"].append(labels[i])
    
print(data)
```

```python
import seaborn as sns

sns.set_theme(style="ticks")

g = sns.scatterplot(data=data, x="x1", y="x2", hue="label", palette={1:"red", -1:"blue"}, s=300)
g.set_xticks([-1, 0, 1])
g.set_xticklabels([-1, 0, 1])
g.set_yticks([0, 1])
g.set_yticklabels([0, 1])
```

```python
import numpy as np
import matplotlib.pyplot as plt

g = sns.scatterplot(data=data, x="x1", y="x2", hue="label", palette={1:"red", -1:"blue"}, s=300)
g.set_xticks([-1, 0, 1])
g.set_xticklabels([-1, 0, 1])
g.set_yticks([0, 1])
g.set_yticklabels([0, 1])

for i in range(len(data['x1'])):
    g.add_patch(plt.Circle(xy=(data['x1'][i], data['x2'][i]), radius=0.3, color='red', fill=False))

w = np.array([0, -2])
X_plot = np.linspace(-1, 1)
Y_plot = np.array([np.dot(w, phi(x)) + 1 for x in X_plot])

g.plot(X_plot, Y_plot, color='g')
```
